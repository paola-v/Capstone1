<h2><center> Combine the main dataframes to work with and do some aggregations and group operations</center></h2>

<h4> Merge product and unit price tables based on prod_id key and drop unwanted columns <h4>
<br>

```python
pu_merged = pd.merge(product_df,unit_price_df, how="inner",on="prod_id")
```

```python
pu_merged_df=pu_merged.drop(['manufacturer','url'],axis=1)
```

```python
pu_merged_df.head()
```

<h4> Merge this new combined product and unit price dataframe with sales and drop unwanted columns </h4>
<br>

```python
sp_merged = pd.merge(sales_df,pu_merged_df,how="inner",on="item_code_year_quarter",suffixes=('','_drop'))
```

```python
sp_merged.drop([col for col in sp_merged.columns if 'drop' in col],axis=1,inplace=True)
```

```python
sp_merged_df = sp_merged.drop(['item_code_year_quarter','prod_id','sales_period_id'],axis=1)
```

Re-order the columns in a more user friendly way of this main dataframe

```python
sp_merged_df = sp_merged_df[['sales_id','emp_id','item_code','prod_name','quantity','price','sales_year','sales_quarter','sales_week']]
```

```python
sp_merged_df
```

```python
sp_merged_df['revenue'] = sp_merged_df['quantity']*sp_merged_df['price']
```

```python
sp_merged_df
```

<h4>Show total revenue by sales team</h4>
<br>

```python
sp_emp = sp_merged_df['revenue'].groupby(sp_merged_df['emp_id'])
```

```python
sp_emp.sum()
```

<h4> Show total revenue by product</h4>
<br>

```python
sp_item = sp_merged_df['revenue'].groupby(sp_merged_df['item_code'])
```

```python
sp_item.sum()
```

<h4>Show total revenue of each product sold by each sales team</h4>
<br>

```python
sp_emp_item = sp_merged_df['revenue'].groupby([sp_merged_df['item_code'],sp_merged_df['emp_id']])
```

```python
sp_emp_item.sum()
```

<h4> Create a pivot table to show the sales of each product by sales team borken down by sales years and sales quarters</h4>
<br>

```python
sp_merged_df.pivot_table('revenue',index=['sales_year','sales_quarter'],
                         columns=['item_code','emp_id'],aggfunc='sum',fill_value=0)
```
